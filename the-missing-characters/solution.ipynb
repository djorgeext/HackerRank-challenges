```python
import nltk
from nltk.corpus import gutenberg
from nltk.util import bigrams
from collections import Counter, defaultdict
import string
```

```python
nltk.download('gutenberg')
nltk.download('punkt')

training_words = [w.lower() for w in gutenberg.words()]
```

```python
# Create a dictionary of word counts (for unigrams) and bigram counts
word_counts = Counter(training_words)
bigram_counts = defaultdict(Counter)

# Populate our bigram memory: bigram_counts[word1][word2]
for w1, w2 in bigrams(training_words):
    bigram_counts[w1][w2] += 1
    
alphabet = "abcdefghijklmnopqrstuvwxyz"

sample_input = """ William Shakespea#e was an Englis# poet, playwright a#d actor, widely regarded as the greatest writer in the Eng#ish langu#ge and the worl#'s pre-eminent dramat#st. He is ofte# called Englan#'s national poet an# the "Bard of Avon".
"""
sample_input = sample_input.lower().split(' ')
words_with_missing_characters = [word.lower() for word in sample_input if '#' in word]
```

```python
missing_letters = []

for i in range(len(words_with_missing_characters)):
    
    idx = sample_input.index(words_with_missing_characters[i])
    context_word = sample_input[idx - 1]

    # 3. Generate all possible real-word candidates by replacing '#' with a-z
    candidates = []
    for letter in string.ascii_lowercase:
        possible_word = words_with_missing_characters[i].replace('#', letter)
        # Only keep it if it's a real English word that exists in our corpus
        if possible_word in word_counts:
            candidates.append(possible_word)
    
    # 4. Score candidates based on how often they follow "playwright"
    best_word = None
    highest_score = -1

    for candidate in candidates:
        # How often does "playwright [candidate]" appear?
        score = bigram_counts[context_word][candidate]
        
        if score > highest_score:
            highest_score = score
            best_word = candidate
    

    # 5. Output the result
    if best_word:
        sample_input[idx] = best_word
        idx_missing = words_with_missing_characters[i].index('#')
        missing_letters.append(best_word[idx_missing])
```
